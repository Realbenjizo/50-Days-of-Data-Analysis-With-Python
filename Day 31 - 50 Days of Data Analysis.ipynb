## 1.	

#### Import the employees_record_data dataset and save it as a table in SQL. Write code to fetch the table, first using fetchall() and then read_sql_query().

```python
import pandas as pd

df = pd.read_csv("employees_record_data.csv")
df.head()
```

```python
import sqlite3

# Connect to SQL and create a database
engine = sqlite3.connect('Employees_Records.db')
cur = engine.cursor()

# Save to SQL
try:
    df.to_sql('MyEmployeesTable', con=engine, index = False)
except ValueError:
    print("Table already exists")
```

```python
df_data = cur.execute("""SELECT * FROM MyEmployeesTable""")
df_data.fetchall()
```

```python
data_table = pd.read_sql_query("""
    SELECT * FROM MyEmployeesTable
   
     """, con=engine)

data_table.head()
```

## 2.	

#### Using pandas sql_query, what is the names and salaries of employees that make over $60,000

```python
salary_over_60k = pd.read_sql_query("""
    SELECT Name, Salary
    FROM MyEmployeesTable
    WHERE Salary > 60000
""", con=engine)

salary_over_60k
```

## 3.	

#### Write a code to return the average salary of each department. Group it by department

```python
ave_salary = pd.read_sql_query("""
    SELECT Department, AVG(Salary) AS Average_salary
    FROM MyEmployeesTable
    GROUP BY Department
""", con=engine)

ave_salary
```

## 4.	

#### Which department has the highest-paid employee? What is their name?

```python
highest_salary = pd.read_sql_query("""
    SELECT Department, Name, MAX(Salary) AS Highest_salary
    FROM MyEmployeesTable
""", con=engine)

highest_salary
```

#### Which employees where hired before 2020?

```python
hire_date_before_2020 = pd.read_sql_query("""
    SELECT Name
    FROM MyEmployeesTable
    WHERE HireDate < "2020-01-01"
""", con=engine)

hire_date_before_2020
```
