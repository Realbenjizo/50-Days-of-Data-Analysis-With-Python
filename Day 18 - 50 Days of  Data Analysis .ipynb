## 1.

#### Import the running_data using pandas

```python
import pandas as pd

df = pd.read_csv("running_data.csv")
df.head()
```

## 2. 

#### Create a copy of the DataFrame. Add another column to the DataFrame that will convert the kilometers run into miles.

```python
# Creating a copy of the DataFrame
df_copy = df.copy()

# Converting kms to miles and creating a new column
df_copy["Distance (Mi)"] = df_copy["Distance (km)"] / 1.60934
df_copy.head()
```

## 3. 

#### Add another column to the DataFrame to calculate the speed per hour in kilometers.

```python
# Calculating the speed in kms
df_copy["speed_in_kms(hr)"] = df_copy["Distance (km)"] / df_copy["Time (hours)"]
df_copy.head()
```

## 4. 

#### 	You have come across some more information to be added to the DataFrame. Some runners stopped for some rests. This information is saved in a CSV dataset called "runner_rest"

#### Merge this dataset to your DataFrame above. Use the pandas merge() method. Ensure to drop the "Runners" column since you already have the "Names" column.

```python
df2 = pd.read_csv("runner_rest.csv")
df2.head()
```

```python
# New available information
df2 = pd.read_csv("runner_rest.csv")

# Merge the two DataFrames 
df_merged = df_copy.merge(df2, left_on="Name", 
                          right_on="Runners", 
                          how="left").drop(columns=["Runners"])

df_merged.head()
```

## 5.

#### Convert the "rest time in min" column into hours using the apply function. Add a new column called "rest time in hours".

```python
# Converting mins to hrs and adding a new column
df_merged["Rest Time(Hrs)"] = df_merged["Rest Time (Mins)"].apply(lambda x:
                                                      x/60 if x is not None else None) 
                                                                  
df_merged.head()
```

## 6

#### Using the pandas map series, create a subset of the DataFrame with runners that covered over 5 miles. Reset the index of the resulting Dataframe. Which runner covered the most distance?

```python
over_5_miles_run = df_merged[df_merged["Distance (Mi)"]
                             .map(lambda x: x) > 5].reset_index(drop=True)
over_5_miles_run.head()
```

```python
# retrieving the index with max miles
max_distance_index = over_5_miles_run["Distance (Mi)"].idxmax()

# Using index to retrieve name with max miles
name = over_5_miles_run["Name"][max_distance_index]
print("The longest distance was achieved by", name)
```
