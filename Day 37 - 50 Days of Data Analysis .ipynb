## 1. 
#### Load the data set. Write a code to check the data types of the columns in the dataset. Check for missing data.

```python
import pandas as pd

df = pd.read_csv("fusion_stock_data.csv")
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

## 2.

#### Convert the date column into datetime format and set it as the index.

```python
# Convert data types into date time
df["Date"] = pd.to_datetime(df["Date"])

# Set the date as the index
df.set_index("Date", inplace=True)
df.head()
```

## 3. 

#### Using Matplotlib and pandas analyze the stock price over time in the stock_data dataset.

```python
import matplotlib.pyplot as plt

df.plot(y='Price', figsize=(10,8))
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price Over Time', fontsize = 20)
plt.show()
```

## 4. 

#### What was the volume on the day with the lowest stock price?

```python
# Using loc to find the volume on lowest price
volume_lowest_price = df.loc[df['Price'].idxmin()]["Volume"]

print(f'Volume on lowest price: {volume_lowest_price}')
```

## 5. 

#### Calculate the daily returns of the stock price using pandas 'pct_change'. Plot a line plot using pandas.

```python
df["return"] = df["Price"].pct_change()


df.plot(y='return', figsize=(10,8))
plt.xlabel('Date')
plt.ylabel('Return')
plt.title('Daily Returns of Stock Price', fontsize = 20)
plt.show()
```

## 6. 

#### What date had the highest price?

```python
# Finding the maximum price in the price column
max_price = df.loc[:, "Price"].max()

# Returning row with maximum price
max_price_row = df.loc[df["Price"] == max_price]

# Retrieving the date
date_with_max_price = max_price_row.index[0]
date_with_max_price
```
