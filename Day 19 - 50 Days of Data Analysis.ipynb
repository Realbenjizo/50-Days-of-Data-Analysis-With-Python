## 1. 

#### Create a DataFrame from the lists above using pandas

```python
import pandas as pd

names = ["Joe", "Phil", "Ken", "Jos", 'Luke']
miles_run = [120, 80, 100, 90, 85]
time_in_hrs = [40, 38, 45, 50, 50]


df = pd.DataFrame({"Names": names, "Miles run": miles_run,
                         "Time in hrs": time_in_hrs})
df
```

## 2. 

#### More information has become available below. Create a new DataFrame with this information. Merge it to the DataFrame above using pandas concat() function.

```python
# New information
names = ["Joe", "Phil", "Ken", "Jos", 'Luke']
Age = [45, 28, 21, 55, 62]
gender = ["M", "M","F", "F", "M"]

# Creating a new DataFrame.
new_df = pd.DataFrame({'Names': names, 'Age': Age, 'Gender': gender})

# Concatenate new_df to old_df and dropping the Names column
df = pd.concat([df, new_df.drop(columns="Names")], axis=1)

df
```

## 3. 

#### Shuffling DataFrames can randomize the data. Write a code to reshuffle the index of the DataFrame.

```python
# Reshuffling the dataset
df = df.sample(frac=1)
df
```

## 5. 

#### Using pandas melt() function reshape your DataFrame to long format. The "Names" column must remain in place.

```python
df_set_long = df.melt(id_vars=["Names"], 
        var_name="Variable", 
        value_name="Values")

df_set_long
```
