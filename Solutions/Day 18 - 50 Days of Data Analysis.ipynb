## 1.

#### Create a DataFrame from the data above.

```python
import pandas as pd

names = ["Joe", "Phil", "Ken", "Jos", 'Luke']
miles_run = [120, 80, 100, 90, 85]
time_in_hrs = [40, 38, 45, 50, 50]


df = pd.DataFrame({"names": names, 
                         "miles_run": miles_run, 
                         "time_in_hrs": time_in_hrs})
df
```

## 2. 

#### Add another column to the DataFrame that will convert the miles run into kms.

```python
# Converting kms to miles and creating a new column
df["kms_run"] = df["miles_run"] * 1.60934
df
```

## 3. 

#### Create another column that has speed run per hour in kms.

```python
# Calculating the speed in kms
df["speed_in_kms_per_hr"] = df["kms_run"] / df["time_in_hrs"]
df
```

## 4. 

#### You have come across some more information to be added to the DataFrame. Some runners stopped for some rests:

#### Add a column called "rest time in min" to the data frame with this information. Use the pandas merge() function. Ensure to drop the "runners" column since you already have the "names" column.

```python
# New available information
runners = ["Joe","Ken", "Jos"]
rest_time = [30, 50, 35]

# creating a new DataFrame with the addition information
df2 = pd.DataFrame({"runners": runners, 
                          "rest_time_in_min": rest_time})

# Merge the two DataFrames 
df = df.merge(df2, left_on="names", 
                          right_on="runners", 
                          how="left").drop(columns=["runners"])

df
```

## 5.

#### Convert the "rest time in min" column into hours using the apply function. Add a new column called "rest time in hours".

```python
# Converting mins to hrs and adding a new column
df["rest_time_in_hrs"] = df["rest_time_in_min"].apply(lambda x:
                                                      x/60 if x is not None else None) 
                                                                  
df
```

## 6

#### Using the pandas map series, create a subset of the DataFrame with runners that covered over 80 miles

```python
over_80_miles_run = df[df['miles_run'].map(lambda x: x) > 80]
over_80_miles_run
```
